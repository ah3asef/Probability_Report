```python
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import skew, kurtosis
```

```python
data = pd.read_csv("C:/College Semester 3/Probability II/Assignments & Solutions/skewed_data.csv",encoding='ISO-8859-1')
```

```python
data_2 = data.copy()
```

```python
data.describe()
```

```python
plt.figure(figsize=(15, 5))
for i, (data, title) in enumerate(zip(
    [data['Positively Skewed'], data['Negatively Skewed'],data['Normal Distribution']],
    ["Positive Skew", "Negative Skew", "Symmetrical"]
)):
    plt.subplot(1, 3, i+1)
    sns.histplot(data, kde=True, bins=30)
    plt.title(f"{title}\nSkewness: {skew(data):.2f}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
