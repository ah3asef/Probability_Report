```python
plt.figure(figsize=(15, 5))
for i, (data, title) in enumerate(zip(
    [data['Positively Skewed'], data['Negatively Skewed'],data['Normal Distribution']],
    ["Positive Skew", "Negative Skew", "Symmetrical"]
)):
    plt.subplot(1, 3, i+1)
    sns.histplot(data, kde=True, bins=30)
    plt.title(f"{title}\nSkewness: {skew(data):.2f}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python

data = [23, 25, 27, 32, 35, 22, 37, 39, 31, 34, 21, 28, 25, 24]
stems = [num // 10 for num in data]
leaves = [num % 10 for num in data]
stem_unique = sorted(set(stems))
leaf_counts = [stems.count(stem) for stem in stem_unique]
plt.figure(figsize=(8, 5))
plt.bar(stem_unique, leaf_counts, color="skyblue", edgecolor="black", width=0.5)
plt.xticks(stem_unique, labels=[f"{stem}" for stem in stem_unique])
plt.title("Stem-and-Leaf Plot (Visualized as Bar Chart)", fontsize=14)
plt.xlabel("Stem (Tens Place)", fontsize=12)
plt.ylabel("Frequency of Leaves", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

stems = [num // 10 for num in data_3['Random Numbers']]#calculating the leading digit and put it in list
leaves = [num % 10 for num in data_3['Random Numbers']]#calculating the last digit and put it in list

stem_unique = sorted(set(stems))#sort the leading digits
leaf_counts = [stems.count(stem) for stem in stem_unique]#count the frequency of each digit


plt.figure(figsize=(8, 5))
plt.bar(stem_unique, leaf_counts, color="skyblue", edgecolor="black", width=0.5)
plt.xticks(stem_unique, labels=[f"{stem}" for stem in stem_unique])
plt.title("Stem-and-Leaf Plot ", fontsize=14)
plt.xlabel("Stem digits", fontsize=12)
plt.ylabel("Frequency of Leaves", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
